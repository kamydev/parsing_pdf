# Basic OCR on Cropped Table

```python
import cv2
import numpy as np
import tensorflow as tf
from paddleocr import PaddleOCR

ocr = PaddleOCR(lang='fr')  # or use_textline_orientation=True if needed

def extract_table_text(image_path):
    image = cv2.imread(image_path)
    result = ocr.predict(image_path)[0]

    if not result:
        print("‚ùå No OCR results.")
        return

    boxes = []
    texts = []
    scores = []

    for line in result:
        try:
            box = line[0]
            text = line[1][0]
            score = line[1][1]

            boxes.append(box)
            texts.append(text)
            scores.append(score)
        except (IndexError, TypeError):
            print(f"‚ö†Ô∏è Skipped malformed line: {line}")
            continue

    if not boxes:
        print("‚ùå No valid OCR results.")
        return

    image_height, image_width = image.shape[:2]
    horiz_boxes = []
    vert_boxes = []

    for box in boxes:
        x_min = int(box[0][0])
        y_min = int(box[0][1])
        x_max = int(box[2][0])
        y_max = int(box[2][1])

        horiz_boxes.append([0, y_min, image_width, y_max])
        vert_boxes.append([x_min, 0, x_max, image_height])

    horiz_nms = tf.image.non_max_suppression(horiz_boxes, scores, max_output_size=100, iou_threshold=0.1).numpy()
    vert_nms = tf.image.non_max_suppression(vert_boxes, scores, max_output_size=100, iou_threshold=0.1).numpy()

    horiz_lines = sorted(horiz_nms)
    vert_lines = sorted(vert_nms)

    print("üßæ Extracted Text Lines:")
    for t in texts:
        print(t)

    print(f"\nüìè Horizontal Lines Detected: {len(horiz_lines)}")
    print(f"üìê Vertical Lines Detected: {len(vert_lines)}")

    return texts, horiz_lines, vert_lines
```

```python
extract_table_text("./output/page_table_1_1.png")
```

```python
# Simple Table OCR Test - Clean Version
import cv2
import numpy as np
import tensorflow as tf
from paddleocr import PaddleOCR

# Configuration
IMAGE_PATH = "./extracted_table.jpg"  # Change this to your image path

# Initialize OCR
print("Loading OCR model...")
ocr = PaddleOCR(lang='fr')
print("‚úì OCR model loaded successfully")

# Load image
print("Loading image...")
image = cv2.imread(IMAGE_PATH)
if image is None:
    print(f"‚ùå Could not load image: {IMAGE_PATH}")
else:
    print(f"‚úì Image loaded: {image.shape}")
    
    # Save as extracted table for OCR
    cv2.imwrite('extracted_table.jpg', image)
    print("Image saved as 'extracted_table.jpg'")

    # Run OCR
    print("Running OCR...")
    result = ocr.ocr('extracted_table.jpg')
    
    # Handle OCR results safely
    if result:
        print(f"OCR result type: {type(result)}")
        print(f"OCR result length: {len(result) if hasattr(result, '__len__') else 'No length'}")
        print(f"First few items: {result[:2] if hasattr(result, '__getitem__') else 'Not indexable'}")
        
        # Try to access the actual data
        try:
            if isinstance(result, list) and len(result) > 0:
                ocr_data = result[0]
                if ocr_data is not None:
                    print(f"OCR data type: {type(ocr_data)}")
                    print(f"OCR data length: {len(ocr_data) if hasattr(ocr_data, '__len__') else 'No length'}")
                    
                    if isinstance(ocr_data, list):
                        print(f"‚úÖ OCR SUCCESS! Found {len(ocr_data)} text elements:")
                        
                        # Show first few results
                        for i in range(min(5, len(ocr_data))):
                            line = ocr_data[i]
                            print(f"  Line {i}: {line}")
                            try:
                                text = line[1][0]
                                confidence = line[1][1]
                                print(f"    Text: '{text}' (conf: {confidence:.3f})")
                            except Exception as e:
                                print(f"    Error parsing line: {e}")
                    else:
                        print(f"OCR data is not a list: {ocr_data}")
                else:
                    print("OCR data is None")
            else:
                print(f"Result is not a proper list: {result}")
        except Exception as e:
            print(f"Error accessing OCR result: {e}")
            import traceback
            traceback.print_exc()
        
        else:
            print("‚ùå No OCR results found")
```
