```python
from paddleocr import PaddleOCR
import cv2
import numpy as np

# Always use this form, as per your instruction
ocr = PaddleOCR(lang='fr')

# Load the image
img_path = "./output/page_table_1_1.png"
image = cv2.imread(img_path)
if image is None:
    raise FileNotFoundError(f"Image not found: {img_path}")
image_height, image_width = image.shape[:2]

# Run OCR
result = ocr.predict(img_path)[0]  # Extract first result dict

# Extract raw text and box data
boxes = result['rec_boxes']
texts = result['rec_texts']

# Defensive check
if len(boxes) != len(texts):
    raise ValueError("Mismatch between number of boxes and texts")

# Handle simple boxes: [x1, y1, x2, y2]
def get_y_center(box):
    x1, y1, x2, y2 = box
    return (y1 + y2) / 2

def get_x_center(box):
    x1, y1, x2, y2 = box
    return (x1 + x2) / 2

# Zip and sort by vertical position
box_texts = [(boxes[i], texts[i]) for i in range(len(boxes)) if isinstance(texts[i], str)]
box_texts.sort(key=lambda x: get_y_center(x[0]))

# Group into rows by y distance
rows = []
current_row = []
row_threshold = 20  # pixel distance between lines

for box, text in box_texts:
    y = get_y_center(box)
    if not current_row or abs(y - get_y_center(current_row[-1][0])) <= row_threshold:
        current_row.append((box, text))
    else:
        current_row.sort(key=lambda x: get_x_center(x[0]))
        rows.append([t for b, t in current_row])
        current_row = [(box, text)]

# Add last row
if current_row:
    current_row.sort(key=lambda x: get_x_center(x[0]))
    rows.append([t for b, t in current_row])

# Output the table
for row in rows:
    print(row)
```
